```python
#!pip install rdflib
import rdflib
from rdflib import Graph
import pandas as pd
import json
import urllib.parse
```

```python
g = Graph()
g.parse("test_data_from_spaqrl_onlysub.ttl")
```

```python
df = pd.read_csv("top40_old.csv", header=None)
```

```python
top40 = df[0].tolist()
```

```python
for i in range(len(top40)):
    top40[i] = rdflib.term.URIRef(top40[i])
```

```python
query = """
SELECT ?s ?p ?o
WHERE {
  ?s ?p ?o .
}
"""

bad_rule_media_test = []

filtered_gq = g.query(query, initBindings = {
    "p": rdflib.term.URIRef("https://w3id.org/biolink/vocab/occurs_in")
})
for row in filtered_gq:
    if row.o not in top40 and row.s[0:40] == "http://purl.obolibrary.org/obo/NCBITaxon" and row.o not in bad_rule_media_test:
      bad_rule_media_test.append(row.o)

```

```python
len(bad_rule_media_test)
```

```python
query = """
SELECT ?s ?p ?o
WHERE {
  ?s ?p ?o .
}
"""

print(len(g))

for row in bad_rule_media_test:
   filtered_gq = g.query(query, initBindings = {
      "p": rdflib.term.URIRef("https://w3id.org/biolink/vocab/occurs_in"),
      "o": rdflib.term.URIRef(row)
    })
   for triple in filtered_gq:
      g.remove(triple)

print(len(g))
```
