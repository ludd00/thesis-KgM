```python
#!pip install rdflib
import rdflib
from rdflib import Graph
import pandas as pd
import json
import urllib.parse
```

```python
g = Graph()
g.parse("kg-mic.ttl")
```

```python
query = """
SELECT ?s ?p ?o
WHERE {
  ?s ?p <https://w3id.org/biolink/vocab/OrganismTaxon> .
}
"""

taxons = []

filtered_gq = g.query(query)
for row in filtered_gq:
    if row.s[0:40] == "http://purl.obolibrary.org/obo/NCBITaxon":
      taxons.append(row.s)

```

```python
taxons
```

```python
query = """
SELECT ?s ?p ?o
WHERE {
  ?s ?p ?o .
}
"""


for row in taxons:
    filtered_gq = g.query(query, initBindings = {
      "s": row
    })
    for triple in filtered_gq:
        if triple.p == rdflib.term.URIRef("https://w3id.org/biolink/vocab/occurs_in"):
            taxons.remove(row)
            break

len(taxons)

```
